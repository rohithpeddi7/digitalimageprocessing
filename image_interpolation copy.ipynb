# Image Interpolation Experiment

```python
# cv2 for reading image data.
import cv2
import numpy as np
import math
```

```python
given_image = cv2.imread('im1.jpg')
# H is height, W is width and C is number of colour channels.
H,W,C = given_image.shape
print(H,W,C)
```

### Nearest Neighbour Interpolation

```python
logo_nni = np.zeros((int(1.5*H),int(1.5*W),C))
for i in range(int(1.5*H)):
    for j in range(int(1.5*W)):
        x_to_take, y_to_take = round(2*i/3),round(2*j/3)
        if x_to_take>=H:
            x_to_take=H-1
        if y_to_take>=W:
            y_to_take=W-1
        logo_nni[i][j][:] = given_image[x_to_take][y_to_take][:]
```

```python
cv2.imwrite('im2.png', logo_nni) 
```

### Bilinear Interpolation

```python
bilinear_logo = np.zeros((1200,1200,C))
for c in range(C):
    for i in range(1200):
        for j in range(1200):
            actual_x,actual_y = i*900/1200,j*900/1200
            x1,x2,y1,y2 = math.floor(actual_x),math.ceil(actual_x),math.floor(actual_y),math.ceil(actual_y)
            if x2>=H:
                x2 = H-1
            if y2>=W:
                y2 = W-1
            #Intermediate intensity for (x,y1)(i1) and (x,y2)(i2)
            if x1!=x2:
                i1 = (given_image[x1][y1][c]*(x2-actual_x) + given_image[x2][y1][c]*(actual_x-x1))
                i2 = (given_image[x1][y2][c]*(x2-actual_x) + given_image[x2][y2][c]*(actual_x-x1))
            else:
                i1 = given_image[x1][y1][c]
                i2 = given_image[x1][y2][c]
            # Now again weighted sum.
            if y1!=y2:
                bilinear_logo[i][j][c] = (i1*(y2-actual_y)+i2*(actual_y-y1))
            else:
                bilinear_logo[i][j][c] = i1

cv2.imwrite('logo_bli.png',bilinear_logo)
```
